## 1. OWASP & CWE Mapping:

OWASP (Open Web Application Security Project) es una organización que identifica las principales vulnerabilidades en aplicaciones web. Su lista “Top 10” muestra los riesgos de seguridad más críticos.

CWE (Common Weakness Enumeration) es un catálogo de tipos de debilidades de software, como errores de programación o fallos de seguridad.

CVE (Common Vulnerabilities and Exposures) son vulnerabilidades reales reportadas en sistemas o aplicaciones, cada una con un identificador único.

## 2. The Ecuadorian Context:

La Ley Orgánica de Protección de Datos Personales establece los derechos que tienen las personas sobre el uso de sus datos personales y regula cómo las organizaciones deben tratarlos.

**1. Derecho de acceso**

Permite al titular conocer qué datos personales están siendo tratados y con qué finalidad

**2. Derecho de rectificación y actualización**

Permite corregir datos incorrectos o desactualizados.

**3. Derecho de eliminación**

Permite solicitar la eliminación de los datos cuando ya no sean necesarios o se hayan tratado de forma indebida.

**4. Derecho de oposición**

Permite oponerse al tratamiento de datos en determinadas circunstancias.

**5. Derecho a la portabilidad**

Permite recibir los datos personales en un formato estructurado y transferirlos a otro responsable.

**6. Derecho a la suspensión del tratamiento**

Permite limitar temporalmente el uso de los datos personales.

**7. Derecho a no ser objeto de decisiones automatizadas**

Protege al titular frente a decisiones basadas únicamente en procesos automáticos.

**8. Derecho a la información**

Permite conocer cómo, por qué y para qué se están utilizando los datos personales.

En Ecuador, la Superintendencia de Protección de Datos Personales es la entidad encargada de garantizar el cumplimiento de la Ley Orgánica de Protección de Datos Personales. Su función principal es supervisar que las instituciones públicas y privadas manejen los datos personales de forma adecuada y respeten los derechos de las personas. Además, puede recibir denuncias cuando alguien considera que su información fue usada de manera incorrecta, realizar investigaciones y aplicar sanciones si encuentra incumplimientos. También emite lineamientos para orientar a las organizaciones sobre cómo deben tratar los datos. Sin embargo, como es una ley relativamente nueva en el país, todavía está en proceso de implementación y su cumplimiento aún no es tan fuerte en la práctica.

Por otro lado, el Reglamento General de Protección de Datos (GDPR) es un sistema mucho más desarrollado y con mayor experiencia en su aplicación. En la Unión Europea existen autoridades de control en cada país que trabajan de forma coordinada para asegurar que se cumpla la normativa. Además, el GDPR Además, exige más a las organizaciones. Por ejemplo, deben aplicar medidas de seguridad desde el diseño de sus sistemas (lo que se conoce como privacy by design), nombrar en algunos casos a un responsable de protección de datos (DPO), y notificar obligatoriamente a las autoridades cuando ocurre una filtración de datos personales en un tiempo máximo establecido. También deben pedir consentimiento claro para el uso de datos y garantizar que las personas puedan ejercer fácilmente sus derechos. Todo esto hace que las empresas tengan más responsabilidad y control sobre cómo manejan la información. También incluye sanciones económicas bastante altas, lo que hace que las empresas se tomen más en serio su cumplimiento. En comparación, En Ecuador también se contemplan varias de estas medidas dentro de la ley, como el deber de proteger la información y respetar los derechos del titular. Sin embargo, en Europa estas obligaciones están más desarrolladas y se aplican de forma más estricta, mientras que en Ecuador todavía se están fortaleciendo en la práctica.

## 5. Design - The Privacy Engineering Task

###Aviso de privacidad de QuitoCash

QuitoCash es una aplicación fintech ecuatoriana que permite enviar y recibir dinero usando números de teléfono. También utiliza inteligencia artificial para analizar patrones generales de gasto y ofrecer recomendaciones financieras personalizadas.

La aplicación de QuitoCash trata datos personales conforme a la Ley Orgánica de Protección de Datos Personales de Ecuador (LOPDP) y al Reglamento General de Protección de Datos de la Unión Europea (GDPR) cuando aplica a usuarios europeos.

Se recolectan únicamente los datos necesarios para operar el servicio, como por ejemplo: nombre, número de teléfono, correo electrónico, documento de identidad, información de cuenta, historial de transacciones y datos básicos de uso de la aplicación. Estos datos se usan para crear la cuenta, verificar la identidad del usuario, procesar transferencias, prevenir fraude, cumplir obligaciones legales y mejorar la experiencia dentro de la app.

Las bases legales para el tratamiento son: ejecución de contrato, cuando se usan los datos para prestar el servicio de envío de dinero; cumplimiento de obligaciones legales, especialmente normas financieras, tributarias y de prevención de fraude; interés legítimo, para seguridad, prevención de abuso y mejora del servicio; y consentimiento, para funciones opcionales como recomendaciones personalizadas mediante inteligencia artificial o comunicaciones comerciales. El GDPR reconoce estas bases legales en su artículo 6.

Los usuarios pueden ejercer sus derechos de acceso, rectificación, eliminación, oposición, portabilidad, limitación del tratamiento, información y a no ser sujetos a decisiones automatizadas, derechos reconocidos también en la LOPDP ecuatoriana.

Para ejercer el derecho de portabilidad, el usuario puede solicitar una copia de sus datos en un formato estructurado, común y reutilizable, por ejemplo CSV o JSON. Para ejercer el derecho de oposición, el usuario puede pedir que QuitoCash deje de tratar sus datos para fines de marketing, perfilamiento o recomendaciones automatizadas, salvo que exista una obligación legal o una razón legítima superior.

Sobre la función de inteligencia artificial, QuitoCash analiza patrones de gasto para mostrar recomendaciones financieras. Sin embargo, estas predicciones no tomarán decisiones legales o financieras importantes de forma totalmente automatizada, como bloquear una cuenta, negar un servicio o limitar una transferencia sin revisión humana. El GDPR reconoce el derecho a no ser sujeto a decisiones basadas únicamente en tratamiento automatizado cuando produzcan efectos legales o significativos.

QuitoCash conservará los datos solo durante el tiempo necesario para prestar el servicio y cumplir obligaciones legales. Después de ese periodo, los datos serán eliminados, anonimizados o bloqueados de forma segura. El usuario puede comunicarse con QuitoCash para ejercer sus derechos mediante el correo: privacidad@quitocash.ec
